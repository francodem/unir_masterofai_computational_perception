```python
import cv2 as cv
from google.colab.patches import cv2_imshow
```

```python
# get the image from a path
# CHANGE the PATH by your own image path to read
PATH = "/content/images/DALLÂ·E 2023-11-06 16.44.48 - Create a hyper-realistic image of a diverse team of AI engineers collaborating in a high-tech office environment. The team includes a Middle-Eastern d.png"
img = cv.imread(PATH)

# resize the image
dimensions = (420, 420)
img = cv.resize(img, dimensions)
```
